#  Data Visulaization
# 1. Risk Score Distribution (Histogram + KDE)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('wallet_risk_scores.csv')

plt.figure(figsize=(10, 6))
sns.histplot(df['score'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Wallet Risk Scores')
plt.xlabel('Risk Score')
plt.ylabel('Number of Wallets')
plt.tight_layout()
plt.show()
```

# 2. Boxplot of Scores

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['score'], color='lightgreen')
plt.title('Boxplot of Wallet Risk Scores')
plt.xlabel('Risk Score')
plt.tight_layout()
plt.show()
```

# 3. Top and Bottom 5 Wallet Profiles (Barplot)

```python
top = df.nlargest(5, 'score')
bottom = df.nsmallest(5, 'score')
sample = pd.concat([top, bottom])

plt.figure(figsize=(12, 6))
sns.barplot(x='wallet_id', y='score', data=sample, palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 5 and Bottom 5 Wallet Risk Scores')
plt.xlabel('Wallet ID')
plt.ylabel('Risk Score')
plt.tight_layout()
plt.show()
```
